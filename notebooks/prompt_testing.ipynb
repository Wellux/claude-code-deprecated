# Prompt Testing

Interactive notebook for testing and iterating on prompts.

```python
import asyncio
import os
import sys
sys.path.insert(0, '..')

from src.llm import ClaudeClient, build_request

client = ClaudeClient(api_key=os.environ.get('ANTHROPIC_API_KEY'))
```

```python
# Test a single prompt
async def test_prompt(prompt, system=None, temperature=0.7):
    req = build_request(prompt, system=system, temperature=temperature)
    resp = await client.complete(req)
    print(resp.content)
    print(f'\nTokens: {resp.input_tokens}+{resp.output_tokens} | Cost: ${resp.cost_usd:.5f}')
    return resp

await test_prompt('Explain transformer attention in 2 sentences.')
```

```python
# Compare two prompt variants
PROMPT_A = 'Summarize RAG systems.'
PROMPT_B = 'Summarize RAG systems. Focus on the retrieval step. Max 3 sentences.'

resp_a = await test_prompt(PROMPT_A)
resp_b = await test_prompt(PROMPT_B)

print('\n--- Comparison ---')
print(f'A tokens: {resp_a.output_tokens} | B tokens: {resp_b.output_tokens}')
```
