# Model Experimentation

Compare outputs across models (opus, sonnet, haiku) and temperatures.

```python
import asyncio, os, sys
sys.path.insert(0, '..')
from src.llm import ClaudeClient, build_request

MODELS = [
    'claude-haiku-4-5-20251001',   # Fast, cheap
    'claude-sonnet-4-6',            # Balanced
    'claude-opus-4-6',              # Powerful, expensive
]

client = ClaudeClient()
```

```python
# Compare models on same prompt
PROMPT = 'In one paragraph, explain why attention mechanisms are fundamental to modern LLMs.'

for model in MODELS:
    req = build_request(PROMPT, model=model, max_tokens=256, temperature=0.3)
    resp = await client.complete(req)
    print(f'\n=== {model} ===')
    print(resp.content[:300])
    print(f'Tokens: {resp.output_tokens} | Cost: ${resp.cost_usd:.5f}')
```

```python
# Temperature sweep
TEMPS = [0.0, 0.3, 0.7, 1.0]
CREATIVE_PROMPT = 'Write a one-sentence metaphor for neural networks.'

for temp in TEMPS:
    req = build_request(CREATIVE_PROMPT, temperature=temp, max_tokens=100)
    resp = await client.complete(req)
    print(f'T={temp}: {resp.content.strip()}')
```
